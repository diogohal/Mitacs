```python
import numpy as np
import pandas as pd
import os
```

```python
dfs = {}
folder_path = 'Ontario_Data/Credit_River_Conservation/New_Data/Hydrometric'
for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, sep=',', low_memory=False).iloc[:,:3]
    df = df.rename(columns={'Stage (m)': 'Level'})
    df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%m/%d/%Y %I:%M:%S %p')
    df = df.iloc[:,2:]
    dfs[file] = df
```

```python
for i, dfn in enumerate(dfs):
    if i == 0:
        min = dfs[dfn].Datetime.min()
        max = dfs[dfn].Datetime.max()
    if dfs[dfn].Datetime.min() < min:
        min = dfs[dfn].Datetime.min()
    if dfs[dfn].Datetime.max() > max:
        max = dfs[dfn].Datetime.max()
full_range = pd.date_range(min, end=max, freq='15T')
df_full_range = pd.DataFrame(full_range, columns=['Datetime'])
df_full_range.head(3)
```

```python
for dfn in dfs:
    df = dfs[dfn]
    df = df.rename(columns={'Level': f'{dfn[:-4]}'})
    df_full_range = df_full_range.merge(df, how='left', on='Datetime')
df_full_range
```

```python
df_full_range.to_csv('crc_new_dataset.csv', sep=',', index=False)
```

## Precipitation integration

```python
folder_path = 'Ontario_Data/Credit_River_Conservation/New_Data/Precipitation'
for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, sep=',')

    df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%m/%d/%Y %I:%M:%S %p')
    df = df.rename(columns={'Precipitation (mm)': f'Precipitation_{file[:-4]}'})
    df = df.drop(columns=['Date', 'Time'], axis='columns')
    df_full_range = df_full_range.merge(df, how='left', on='Datetime')
```

```python
df_full_range.to_csv('crc_new_dataset.csv', sep=',', index=False)
```
